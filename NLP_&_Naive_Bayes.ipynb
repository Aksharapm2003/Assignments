```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import string
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import nltk

# Download the 'punkt_tab' resource
nltk.download('punkt_tab')
```

```python
nltk.download('stopwords')
nltk.download('vander_la_movitt')
nltk.download('wordnet')
```

```python
df = pd.read_csv('/content/blogs.csv')
df
```

```python
df.info()
```

```python
def convert_lower(text):
  return text.lower()
def tokenize(text):
  return nltk.word_tokenize(text)
```

```python
stop_words = set(stopwords.words('english'))
def remove_stopwords(tokens):
  return[word for word in tokens if word not in stop_words]
```

```python
df['processed_text'] = df['Data'].apply(convert_lower)
df['processed_text'] = df['processed_text'].apply(tokenize)
df['processed_text'] = df['processed_text'].apply(remove_stopwords)
```

```python
df['processed_text'] = df['processed_text'].apply(lambda x:''.join(x))
```

```python
tfidf = TfidfVectorizer()
```

```python
tfidf_matrix = tfidf.fit_transform(df['processed_text'])
```

```python
tfidf_df=pd.DataFrame(tfidf_matrix.toarray(),columns=tfidf.get_feature_names_out())
```

```python
tfidf_df
```

```python
tfidf_df.info()
```

```python
X = tfidf.fit_transform(df['processed_text'])
Y = df['Labels']
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
```

```python
 x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
nb = MultinomialNB()
```

```python
nb.fit(x_train,y_train)
```

```python
y_pred = nb.predict(x_test)
y_pred
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
classification_report(y_test,y_pred)
```
